# Data Preprocessing

```python
#Check skewness in numeric features
df.skew()
# 14 Out of 23 columns have skewed.
```

```python
#Removing skewness
for i in df.skew().index:
    if df.skew().loc[i] > 0.5:
        df[i]=np.log1p(df[i])
    if df.skew().loc[i] < -0.5:
        df[i]=np.square(df[i])
```

```python
#Check skewness after
df.skew()
# No of columns having skewness reduced from 14 to 3.
```

```python
#CHeck shape
df.shape
```

```python
#Dividing the data into dependent & independent variables
df_x=df.drop(columns=["Attrition"])
df_y=df["Attrition"]
print(df_x.shape,df_y.shape)
```

```python
#Target variable
df_y
```

```python
#Convert target variable in numerics using label encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(df_y)
y=le.transform(df_y)
y
```

```python
#Check shape after encoding
y.shape
```

```python
#Check independent variables
df_x
```

```python
#Encoding indepndent variables using pd.dummies
df_x=pd.get_dummies(df_x,drop_first=True)
```

```python
#Check independent variables after encoding
df_x
```

```python
#Use Standard scaler to transform data to same scale 
from sklearn.preprocessing import StandardScaler
Std_sc=StandardScaler()
x=Std_sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
x
```

```python
#Check target variabe column
df["Attrition"].value_counts()
#Data imbalancing
```

```python
#To avoid data imbalancing we use oversampling
from imblearn.over_sampling import SMOTE
Sm=SMOTE(random_state=2)
x_smp,y_smp=Sm.fit_sample(x,y)
print('After OverSampling,the shape of x: ',x_smp.shape) 
print('After OverSampling,the shape of y: ',y_smp.shape) 

print("After OverSampling,count of label '1': ",sum(y_smp == 1))
print("After OverSampling,count of label '0': ",sum(y_smp == 0))
```

```python
#Reduce the columns
#Principal Component Analysis to improve model performance.
from sklearn.decomposition import PCA
pca=PCA(n_components=35)
pc_x=pca.fit_transform(x_smp)

#Check information retained in data after PCA
print("Percentage of data retained =",(sum(pca.explained_variance_ratio_)*100))
```

```python
#Independent variables from PCA
print(pc_x.shape)
#Dependent variable from Oversampling
print(y_smp.shape)
```

# Check best Random State

```python
#Lets use f1-score,confusion matrix,classification report & AUC_ROC as metrics 
from sklearn.metrics import f1_score,classification_report,confusion_matrix
from sklearn.metrics import plot_roc_curve
from sklearn.model_selection import train_test_split
def maxf1_score(clf,x,y):
    maxf1=0
    rs=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x, y,random_state = r_state,test_size=0.20,stratify=y)
        clf.fit(x_train,y_train)
        y_pred=clf.predict(x_test)
        tmp=f1_score(y_test,y_pred)
        if tmp>maxf1:
            maxf1=tmp
            rs=r_state
    print("Maximum f1_score is at random state :",rs," and it is :",maxf1)
    print("Confusion matrix \n",confusion_matrix(y_test,y_pred))
    print("Classification report \n",classification_report(y_test,y_pred))
    
    #AUC-ROC Curve-Logistic regression
    lg_disp = plot_roc_curve(clf,x_test,y_test)
    plt.title("ROC")
    plt.show()
```

```python
#User defined function for cross-validation
from sklearn.model_selection import cross_val_score
def cross_val(model,x,y):
    cv=cross_val_score(model,x,y,cv=5,scoring="f1")
    print("Mean f1 score for classifier: ",cv.mean())
    print(cv)
```
